```python
#Importing all desired packages
import cv2
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

```python
#Importing image from folder and applying garyscale
img = cv2.imread("1.jpeg")
plt.imshow(img)
#plt.show()
#cv2.imshow("Img",img)      
```

```python
#Demo image Selected
Image(filename='1.jpeg')
```

```python
#Implementing Laplacian of Gaussian
laplacian_var = cv2.Laplacian(img, cv2.CV_64F).var()

#Checking for Laplacian value
if laplacian_var < 5:
    print("Image is blur !")
else:   
    print("Image is Clear !")
```

```python
#Printing the desired value of Laplacian
print("Laplacian Value of the Image is:- ",laplacian_var)
```
